# Gradient Descent Algorithm For Linear Regression

```python
#import all the required library
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

# configuring the size of frame in which our plot will shown to us.
plt.figure(figsize=(12,9))

# setting the theme of plot
plt.style.use('seaborn')

# Preprocessing input data
data = pd.read_csv('data.csv')

X = data.iloc[:,0]
Y = data.iloc[:,1]

plt.scatter(X,Y)
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression')
plt.show()

```

```python
# Building The algorithm into code
m = 0
c = 0
L = 0.0001 # it is our Learning Rate
iteration = 1000 # it is the number of iteration our algo will do.
n = float(len(X))
# There is note point that value of iteration should be greater then the accuracy of your model increases.
# Keeping in mind that the value of learing rate should small.

for i in range(iteration):
    #prediction value of y based on our m & c
    Y_pred = m*X + c
    #Now this is the partial Diffrentiated value of the error function wrt m and c
    D_m = (-2/n) * sum(X * (Y-Y_pred))
    D_c = (-2/n) * sum(Y-Y_pred)
    #Now update the value of m and c
    m = m-L*D_m
    c = c-L*D_c
    
print(m,c)
    
    
```

```python
#Making the prediction
plt.figure(figsize=(12,9))
plt.style.use('seaborn')
Y_pred = m*X+c
plt.scatter(X,Y)
plt.plot([min(X), max(X)],[min(Y_pred), max(Y_pred)], color = 'red')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression')
plt.show()
```
